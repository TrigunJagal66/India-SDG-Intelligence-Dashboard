```python
# A Random Forest model trained on 100+ SDG indicators identified the top predictors of SDG performance.
# so we will predict composite_score_of_sdg_india_index using all indicators as features
```

```python
import pandas as pd
import sklearn
```

```python
master = pd.read_csv("../output/SDG_Master.csv")
```

```python
master.shape
```

```python
y = master["composite_score_of_sdg_india_index"]
```

```python
X = master.drop(
    columns=[
        "state",
        "year",
        "composite_score_of_sdg_india_index"
    ]
)
```

```python
X.head()
```

```python
X.isnull().sum().sum()
```

```python
missing = X.isnull().sum()

missing[missing > 0].sort_values(ascending=False)
```

```python
X_clean = X.copy()
```

```python
drop_cols = [
    "mean_shore_zone_coastal_water_quality_total_nitrogen_tn",
    "mean_shore_zone_coastal_water_quality_biochemical_oxygen_demand_bod",
    "increase_in_area_under_mangroves_percent",
    "average_marine_acidity_ph_measured_at_representative_sampling_stations_in_the_shore_zone",
    "potential_area_available_for_developed_under_aquaculture_percent"
]

X_clean = X_clean.drop(columns=drop_cols)
```

```python
X_clean.shape
```

```python
X_clean = X_clean.fillna(
    X_clean.median(numeric_only=True)
)
```

```python
X_clean.isnull().sum().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_clean,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

mae = mean_absolute_error(y_test, y_pred)

rmse = mean_squared_error(
    y_test,
    y_pred
) ** 0.5

r2 = r2_score(y_test, y_pred)

print("MAE :", round(mae,2))
print("RMSE:", round(rmse,2))
print("R2  :", round(r2,2))
```

```python
importance_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": rf.feature_importances_
})

importance_df = importance_df.sort_values(
    by="Importance",
    ascending=False
)

importance_df.head(20) # we get importance 
```

```python
sdg_cols = [col for col in master.columns
            if col.startswith("sdg_india_index")] # drop cols of sdg_india_index of composite score

X2 = master.drop(
    columns=sdg_cols +
    ["state",
     "year",
     "composite_score_of_sdg_india_index"]
)
```

```python
drop_cols = [
    "mean_shore_zone_coastal_water_quality_total_nitrogen_tn",
    "mean_shore_zone_coastal_water_quality_biochemical_oxygen_demand_bod",
    "increase_in_area_under_mangroves_percent",
    "average_marine_acidity_ph_measured_at_representative_sampling_stations_in_the_shore_zone",
    "potential_area_available_for_developed_under_aquaculture_percent"
]

X2 = X2.drop(columns=drop_cols)
```

```python
X2 = X2.fillna(
    X2.median(numeric_only=True)
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X2,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(
    y_test,
    y_pred
) ** 0.5
r2 = r2_score(y_test, y_pred)
print("MAE :", round(mae,2))
print("RMSE:", round(rmse,2))
print("R2  :", round(r2,2))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    rf,
    X_clean,
    y,
    cv=5,
    scoring="r2"
)

print(scores)
print("Mean R2:", scores.mean())
```

```python
importance_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": rf.feature_importances_
})

importance_df = importance_df.sort_values(
    by="Importance",
    ascending=False
)

importance_df.head(20) # we get importance 
```
