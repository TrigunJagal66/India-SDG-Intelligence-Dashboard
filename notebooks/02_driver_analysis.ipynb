```python
#Understand WHY some states perform better.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
master = pd.read_csv("../output/SDG_Master.csv")

master.head()
```

```python
print(master.shape)
```

```python
top10 = master.nlargest(
    10,
    "composite_score_of_sdg_india_index"
)

bottom10 = master.nsmallest(
    10,
    "composite_score_of_sdg_india_index"
)
```

```python
top10[["state","composite_score_of_sdg_india_index"]]
```

```python
bottom10[["state","composite_score_of_sdg_india_index"]]
```

```python
# Compare High vs Low Performers

comparison = pd.DataFrame({
    "Top10": top10.mean(numeric_only=True),
    "Bottom10": bottom10.mean(numeric_only=True)
})

comparison["Difference"] = (
    comparison["Top10"] -
    comparison["Bottom10"]
)

comparison.head()
```

```python
comparison.shape
```

```python
comparison.sort_values(
    by="Difference",
    ascending=False
).head(20)
```

```python
comparison.sort_values(
    by="Difference",
    ascending=True
).head(20)
```
